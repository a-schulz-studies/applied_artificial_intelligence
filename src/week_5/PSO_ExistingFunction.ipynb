```python
#                                               iA
import sys
from mealpy.bio_based import SMA
from mealpy.evolutionary_based import GA
import numpy as np
from mealpy.swarm_based.PSO import CL_PSO
noJobs = 50
noMachines = 40
machinesCap = 150
jobRange = 10
trunOnPentaly = 100
job_list0 = (np.random.rand(1,noJobs) * jobRange).astype(int)
job_list = np.squeeze(job_list0)
print(job_list)
bestPossibleScore = sum(job_list) + np.ceil(sum(job_list)/machinesCap)*trunOnPentaly
print('Best possible score: ', bestPossibleScore)
input('Start...')
# mC = np.zeros(5)
# print(mC)


def fitness_func(_sol0):

    _sol = np.squeeze(_sol0.astype(int))
    rew = 0
    mC = np.zeros(noMachines)
    # print('Sol:', _sol)
    for i in range(0, _sol.shape[0]):
        _mInd = _sol[i]
        mC[_mInd] = mC[_mInd] + job_list[i]
    # print('machine:', mC)
    # input(' ... ')
    # print(mC, ' -> Max:', max(mC), ' | but max possible:', machinesCap)
    if max(mC) > machinesCap:
        rew = 10000000
    for j in range(0, mC.shape[0]):
        if mC[j] > 0:
            mC[j] = mC[j] + trunOnPentaly
    z = np.sum(mC) + rew
    # if z > 200:
    #     print('Bad')
    #     input('bad')

    return z


def amend_position(solution, lb, ub):
    ## Bring them back to valid range
    solution = np.clip(solution, lb, ub)
    ## Convert float to integer number
    solution_int = solution.astype(int)
    ## If the designed solution is permutation, then need an extra step here
    ## .... Do it here and then return the valid solution
    return solution_int

_lb = [0,] * noJobs
_lb1 = list(map(int, _lb))

_ub = [noMachines-1,] * noJobs
_ub1 = list(map(int, _ub))

problem_dict = {
    "fit_func": fitness_func,
    "lb": np.asarray(_lb1).astype(int),
    "ub": np.asarray(_ub1).astype(int),
    "amend_position": amend_position,
    "minmax": "min",
    "save_population": True,
}



epoch = 1000
pop_size = 300
c_local = 1.2
w_min = 0.3
w_max = 0.8
max_flag = 9
model = CL_PSO(epoch, pop_size, c_local, w_min, w_max, max_flag)
best_position, best_fitness = model.solve(problem_dict)
print(f"Solution: {best_position}, Fitness: {best_fitness}, /Best_possible_score: {bestPossibleScore}")

# ga_model = GA.BaseGA(epoch=1000, pop_size=600, pc=0.85, pm=0.05)
# best_position, best_fitness_value = ga_model.solve(problem_dict)
# print(f"Solution: {best_position}, Fitness: {best_fitness_value}")


_machines = np.zeros(noMachines)
for i in range(0,len(best_position)):
    _machines[best_position[i]] = _machines[best_position[i]] + job_list[i]
print(job_list)
print(_machines)
```
