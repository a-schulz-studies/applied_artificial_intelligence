```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from skimage import color

img = io.imread('Altstadt-Dresden-1180x786.jpg')
plt.imshow(img)

plt.figure()
imgGray = color.rgb2gray(img)
plt.imshow(imgGray, cmap=plt.cm.gray)

plt.figure(figsize=(20,4), dpi=80)
# X_flattened = imgGray.reshape(imgGray.shape[0], -1).T
X_flattened = imgGray.flatten()
plt.plot(X_flattened,'-k')
```

```python
# Bild in NumPy-Array umwandeln
image_array = np.array(img)

# Bildmaße extrahieren
height, width, channels = image_array.shape

# Bild in einen flachen Vektor umwandeln
flattened_image = image_array.reshape(height * width * channels)

# Ergebnis anzeigen
print(flattened_image.shape)  # Ausgabe: (Flache Länge des Bildes,)
print(flattened_image)
```
