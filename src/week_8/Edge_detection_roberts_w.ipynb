```python
import matplotlib.pyplot as plt
from skimage import data, filters, color

# Load the image (you can replace this with your own image)
image0 = plt.imread('Images/Altstadt-Dresden-1180x786.jpg')

image= color.rgb2gray(image0)

# Apply Roberts edge detection
edges = filters.roberts(image)

# Display the original image and the detected edges
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax0, ax1 = axes.ravel()

ax0.imshow(image, cmap=plt.cm.gray)
ax0.set_title('Original Image')

ax1.imshow(edges, cmap=plt.cm.gray)
ax1.set_title('Roberts Edges')

plt.tight_layout()
plt.show()
```
