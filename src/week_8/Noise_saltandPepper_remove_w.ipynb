```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.color import rgb2gray

# img = cv2.imread("csm_20190130_Industrie4_0-181_Webseite_6051ef008a.jpg",0)
img = cv2.imread("csm_20190130_Industrie4_0-181_Webseite_6051ef008a.jpg", cv2.IMREAD_GRAYSCALE)
# color
# img = cv2.imread("csm_20190130_Industrie4_0-181_Webseite_6051ef008a.jpg")

mean = 0
stddev = 90
noise = np.zeros(img.shape, np.uint8)
cv2.randn(noise, mean, stddev)

# Add noise to image
gn_img = cv2.add(img, noise)

# Save noisy image
# cv2.imwrite('noisy_img.jpg', noisy_img)


denoised1=cv2.fastNlMeansDenoising(gn_img,None,30,30)

fig=plt.figure(dpi=300)
fig.add_subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(gn_img,cmap='gray')
plt.axis("off")
plt.title("with Gaussian Noise")

fig.add_subplot(1,3,3)
plt.imshow(denoised1,cmap='gray')
plt.axis("off")
plt.title("After Denoising")
```
